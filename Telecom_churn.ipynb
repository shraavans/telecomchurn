### Deriving new features and filtering high value customers

Accoriding to the data dictionary months 6,7 are good phase, 8 is action phase and 9 is churn phase. 
Different months have different recharge amounts for data and calls. We will club this and derive the average recharge during the good phase. 

```python
## next we multiply the data total_rech_data with average recharge amount for each month.
months=[6,7,8]

for i in months:
    tele[f'total_data_amt_{i}']=tele[f'total_rech_data_{i}']*tele[f'av_rech_amt_data_{i}']
```

```python
## total charges for each month

for i in months:
    tele[f'total_monthly_{i}']=tele[f'total_data_amt_{i}']+tele[f'total_rech_amt_{i}']
```

```python
### Next we need to find the average recharge amount for the good phase (month 6 and 7)


tele['good_phase_av']=  (tele['total_monthly_6']+ tele['total_monthly_7'])/2

```

```python
tele.shape
```

```python
## filtering high vlaue customers according to the problem statement. 
filter_customers=tele['good_phase_av'].quantile(0.7)
filter_customers
```

```python
#filtering high value customers above 70th percentile.

hvc=tele[tele['good_phase_av']>filter_customers]
```

```python
def model(X_train):
    X_train_sm=sm.add_constant(X_train)
    logm=sm.GLM(y_train,X_train_sm,family=sm.families.Binomial())
    res=logm.fit()
    return res.summary()
```

```python
def VIF(X_train):
    vif=pd.DataFrame()
    vif['Features']=X_train.columns
    vif['VIF Value']=[variance_inflation_factor(X_train.values,i) for i in range(X_train.shape[1])]
    vif['VIF Value']=round(vif['VIF Value'],2)
    vif=vif.sort_values(by='VIF Value',ascending=False)
    return vif
```

```python
# ROC function
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```

```python
fpr, tpr, thresholds = metrics.roc_curve( y_train_pred_final.churn, y_train_pred_final.churn_probability, drop_intermediate = False )
```

```python
cutoff=pd.DataFrame(columns=['Probability','Accuracy','Sensitivity','Specificity'])
for i in numbers:
    confusion_cutoff=confusion_matrix(y_train_pred_final.churn,y_train_pred_final[i])
    TN=confusion_cutoff[0,0] #True Negative
    TP=confusion_cutoff[1,1] #True Positive
    FN=confusion_cutoff[1,0] #False Negative
    FP=confusion_cutoff[0,1] 
    Accuracy=accuracy_score(y_train_pred_final.churn,y_train_pred_final[i])
    Sensitivity=TP/(TP+FN)
    Specificity=TN/(TN+FP)
    cutoff.loc[i]=[i,Accuracy,Sensitivity,Specificity]
cutoff
```
